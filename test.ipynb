```python
from ultralytics.ultralytics import YOLO
import cv2

# 1. 모델 로드 (사전 학습된 YOLOv8 모델)
try:
    model = YOLO("yolov8n.pt")
    print("✅ YOLOv8 모델 로드 성공")
except Exception as e:
    print(f"❌ 모델 로드 실패: {e}")

# 2. 테스트 이미지 다운로드
image_url = "https://ultralytics.com/images/bus.jpg"
image_path = "test_image.jpg"

try:
    import urllib.request
    urllib.request.urlretrieve(image_url, image_path)
    print("✅ 테스트 이미지 다운로드 성공")
except Exception as e:
    print(f"❌ 테스트 이미지 다운로드 실패: {e}")

# 3. 이미지 예측
try:
    results = model.predict(image_path)
    print("✅ 예측 성공")

    # 4. 예측 결과 시각화
    results.show()

    # 5. matplotlib로 예측 이미지 표시
    img = cv2.imread(image_path)
    if img is not None:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title("YOLOv8 Prediction")
        plt.axis("off")
        plt.show()
    else:
        print("❌ 이미지 로딩 실패 (OpenCV)")
except Exception as e:
    print(f"❌ 예측 실패: {e}")
```
